```python
#read bright and faint data
desy3b = sacc.Sacc.load_fits('../../../TXPipe/data/desy3/Debug/GTSTARS_NoRandoms/gammat_bright_stars.sacc')
desy3f = sacc.Sacc.load_fits('../../../TXPipe/data/desy3/Debug/GTSTARS_NoRandoms/gammat_dim_stars.sacc')
dy3btx = desy3b.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
dy3ftx = desy3f.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
derrb = np.array(desy3b.get_tag('error'))
derrf = np.array(desy3f.get_tag('error'))

desy3b = sacc.Sacc.load_fits('../../../TXPipe/data/desy3/Debug/GTSTARS_RandomsNoWeight//gammat_bright_stars.sacc')
desy3f = sacc.Sacc.load_fits('../../../TXPipe/data/desy3/Debug/GTSTARS_RandomsNoWeight//gammat_dim_stars.sacc')
dy3btxRNW = desy3b.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
dy3ftxRNW = desy3f.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
derrbRNW = np.array(desy3b.get_tag('error'))
derrfRNW= np.array(desy3f.get_tag('error'))

desy3b = sacc.Sacc.load_fits('../../../TXPipe/data/desy3/Debug/GTSTARS_RandomsNGalWeight//gammat_bright_stars.sacc')
desy3f = sacc.Sacc.load_fits('../../../TXPipe/data/desy3/Debug/GTSTARS_RandomsNGalWeight//gammat_dim_stars.sacc')
dy3btxRNGW = desy3b.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
dy3ftxRNGW = desy3f.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
derrbRNGW = np.array(desy3b.get_tag('error'))
derrfRNGW = np.array(desy3f.get_tag('error'))
```

```python
kidsb = sacc.Sacc.load_fits('../../../TXPipe/data/kids/FINALRUN_v2/GTSTARS_NoRandoms/gammat_bright_stars.sacc')
kidsf = sacc.Sacc.load_fits('../../../TXPipe/data/kids/FINALRUN_v2/GTSTARS_NoRandoms/gammat_dim_stars.sacc')
kbtx = kidsb.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
kftx = kidsf.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
kerrb = np.array(kidsb.get_tag('error'))
kerrf = np.array(kidsf.get_tag('error'))

kidsb = sacc.Sacc.load_fits('../../../TXPipe/data/kids/FINALRUN_v2/GTSTARS_RandomsNoWeight/gammat_bright_stars.sacc')
kidsf = sacc.Sacc.load_fits('../../../TXPipe/data/kids/FINALRUN_v2/GTSTARS_RandomsNoWeight/gammat_dim_stars.sacc')
kbtxRNW = kidsb.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
kftxRNW = kidsf.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
kerrbRNW = np.array(kidsb.get_tag('error'))
kerrfRNW = np.array(kidsf.get_tag('error'))

kidsb = sacc.Sacc.load_fits('../../../TXPipe/data/kids/FINALRUN_v2/GTSTARS_RandomsNGalWeight/gammat_bright_stars.sacc')
kidsf = sacc.Sacc.load_fits('../../../TXPipe/data/kids/FINALRUN_v2/GTSTARS_RandomsNGalWeight/gammat_dim_stars.sacc')
kbtxRNGW = kidsb.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
kftxRNGW = kidsf.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
kerrbRNGW = np.array(kidsb.get_tag('error'))
kerrfRNGW = np.array(kidsf.get_tag('error'))
```

```python
hscy3b = sacc.Sacc.load_fits('../../../TXPipe/data/hsc-y3/outputs/GTSTARS/GTSTARS_NoRandoms/gammat_bright_stars.sacc')
hscy3f = sacc.Sacc.load_fits('../../../TXPipe/data/hsc-y3/outputs/GTSTARS/GTSTARS_NoRandoms/gammat_dim_stars.sacc')
hy3btx = hscy3b.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
hy3ftx = hscy3f.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
herrb = np.array(hscy3b.get_tag('error'))
herrf = np.array(hscy3f.get_tag('error'))

hscy3b = sacc.Sacc.load_fits('../../../TXPipe/data/hsc-y3/outputs/GTSTARS/GTSTARS_RandomsNoWeight/gammat_bright_stars.sacc')
hscy3f = sacc.Sacc.load_fits('../../../TXPipe/data/hsc-y3/outputs/GTSTARS/GTSTARS_RandomsNoWeight/gammat_dim_stars.sacc')
hy3btxRNW = hscy3b.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
hy3ftxRNW = hscy3f.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
herrbRNW = np.array(hscy3b.get_tag('error'))
herrfRNW = np.array(hscy3f.get_tag('error'))

hscy3b = sacc.Sacc.load_fits('../../../TXPipe/data/hsc-y3/outputs/GTSTARS/GTSTARS_RandomsNGalWeight/gammat_bright_stars.sacc')
hscy3f = sacc.Sacc.load_fits('../../../TXPipe/data/hsc-y3/outputs/GTSTARS/GTSTARS_RandomsNGalWeight/gammat_dim_stars.sacc')
hy3btxRNGW = hscy3b.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','bright_stars')
hy3ftxRNGW = hscy3f.get_theta_xi('galaxyStar_shearDensity_xi_t','source2d','dim_stars')
herrbRNGW = np.array(hscy3b.get_tag('error'))
herrfRNGW = np.array(hscy3f.get_tag('error'))
```

```python
matplotlib.rcParams.update({'font.size': 12})

fig = plt.figure(figsize=(5.5,5),dpi=100)

gs  = gridspec.GridSpec(2, 1, top=0.98, right=0.96, left=0.16, bottom=0.1,hspace=0.05,wspace=0.15)
ax0 = plt.subplot(gs[0])
ax1 = plt.subplot(gs[1])

ax0.set_ylabel(r'$\gamma_{\rm T bright}/10^{-4}$',labelpad=10)
ax0.errorbar(dy3btx[0],dy3btx[1]/10**-4,derrb/10**-4,marker='o',color=descol,ms=4.5,linestyle='None',label='DES-Y3',capsize=3)
ax0.errorbar(dy3btxRNW[0]+0.1*dy3btxRNW[0],dy3btxRNW[1]/10**-4,derrbRNW/10**-4,marker='o',color='red',ms=4.5,linestyle='None',label='UW Randoms',capsize=3)
ax0.errorbar(dy3btxRNGW[0]-0.1*dy3btxRNGW[0],dy3btxRNGW[1]/10**-4,derrbRNGW/10**-4,marker='o',color='orange',ms=4.5,linestyle='None',label='NG Randoms',capsize=3)

ax0.legend(frameon=False,labelspacing=0.05,ncol=3,handletextpad=-0.2,columnspacing=0.2,loc='upper right',borderpad=0)
ax0.semilogx(np.arange(250)[2:],np.zeros(250)[2:],color='dimgray',ls='--',lw=0.7,dashes=(5,4))
plt.setp(ax0.get_xticklabels(), visible=False)
ax0.tick_params(axis="both",which='both',direction="in")

ax1.set_ylabel(r'$\gamma_{\rm T faint}/10^{-5}$',labelpad=10)
ax1.set_xlabel(r'$\theta [\rm arcmin]$',labelpad=10)
ax1.errorbar(dy3ftx[0],dy3ftx[1]/10**-5,derrf/10**-5,marker='o',color=descol,ms=4.5,linestyle='None',capsize=3)
ax1.errorbar(dy3ftxRNW[0]+0.1*dy3ftxRNW[0],dy3ftxRNW[1]/10**-5,derrfRNW/10**-5,marker='o',color='red',ms=4.5,linestyle='None',capsize=3)
ax1.errorbar(dy3ftxRNGW[0]-0.1*dy3ftxRNGW[0],dy3ftxRNGW[1]/10**-5,derrfRNGW/10**-5,marker='o',color='orange',ms=4.5,linestyle='None',capsize=3)

ax1.semilogx(np.arange(250)[2:],np.zeros(250)[2:],color='dimgray',ls='--',lw=0.7,dashes=(5,4))
ax1.tick_params(axis="both",which='both',direction="in")

from matplotlib.ticker import ScalarFormatter
for axis in [ax1.xaxis]:
    axis.set_major_formatter(ScalarFormatter())
    
for axis in ['top','bottom','left','right']:
    ax0.spines[axis].set_linewidth(1.2)
    ax1.spines[axis].set_linewidth(1.2)
    
#plt.savefig('GTStars.pdf',dpi=200)
```
