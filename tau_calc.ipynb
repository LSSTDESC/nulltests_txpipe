Check tau calculation à la Hamana equation with paper results

```python
tau_ham = h5py.File('./hsc_calcs_external/hsc_tau_hamana.hdf5','r')
tau_hamms = h5py.File('./hsc_calcs_external/hsc_tau_hamanams.hdf5','r')
tau_txham = h5py.File('../TXPipe/data/hscy1/outputsNoEta/tau_stats-hamanajackknife.hdf5') 
tau_txhamms = h5py.File('../TXPipe/data/hscy1/outputsNoEta/tau_stats-hamanamsjackknife.hdf5') 
```

```python
import matplotlib.ticker as ticker

fig = plt.figure(figsize=(6,4.5))
img1 = plt.imread("./hscy1/Hamana2019/Xi_gp.png")

#set first axes
ax = fig.add_subplot(1,1,1)
ax.imshow(img1,  aspect='equal',alpha=0.2)
plt.axis('off')

#create second axes
newax = fig.add_axes(ax.get_position(), frameon=True)
newax.patch.set_alpha(0.0)
newax.set_xlim((2.85,113))

newax.set_ylim((-0.45e-5,2.25e-5))
newax.set_xlabel(r'$\theta\ [{\rm arcmin}]$',size=14)
newax.set_ylabel(r'$\tau_{0+}$',size=15)
newax.set_xscale('log')
#newax.set_yscale('symlog',linthresh=1e-6)

#Change formatting of xticks
newax.xaxis.set_minor_formatter(ticker.FormatStrFormatter('%.1f'))
plt.plot(tau_ham['tau_stats/theta'][:],
         tau_ham['tau_stats/tau0'][:]/2,color="black",linestyle='None',marker='o',label='HSC')

plt.plot(tau_txham['tau_statistics/tau_PSF-reserved/theta'][:],
         tau_txham['tau_statistics/tau_PSF-reserved/tau0'][:]/2,color='black',linestyle='dashed',label='TXPipe')

plt.plot(tau_hamms['tau_stats/theta'][:],
         tau_hamms['tau_stats/tau0'][:]/2,color="brown",linestyle='None',marker='o',label='HSC meansub')

plt.plot(tau_txhamms['tau_statistics/tau_PSF-reserved/theta'][:],
         tau_txhamms['tau_statistics/tau_PSF-reserved/tau0'][:]/2,color='brown',linestyle='dashed',label='TXPipe meansub')

plt.xscale('log')
plt.ylabel(r'$\tau_0$')
plt.xlabel(r'$\theta$')
plt.legend(bbox_to_anchor=(1.05, 1),fontsize=10)
```
